- `validar_senha(senha)` → checa tamanho mínimo + letra + número

## 3. 🔹 Função de criação de conta
Ordem interna:
- Pede nome do personagem → valida

## 4. 🔹 Função de escolha de personagem
- Mostra as opções com descrição temática de cada classe
- Usa `while` pra não aceitar opção inválida

| Opção | Classe | Descrição |
| --- | --- | --- |
| 1 | Mago | Mestre das artes arcanas |

## 5. 🔹 Função de itens iniciais
- Recebe o personagem

## 6. 🔹 Função de evento do baú
- Recebe personagem + inventário
- Adiciona item raro temático da classe

## 7. 🔹 Função de evento do ladrão
- Checa se inventário não está vazio
- Escolhe item aleatório com `random.choice()`
- Remove o item e mostra qual foi roubado com mensagem dramática

## 8. 🔹 Função de exibir ficha final
Mostra um resumo formatado:
```text

=============================
Nome:       Aldric
Classe:     Mago
Email:      aldric@reino.com

```

## 9. 🔹 Função principal `jogar()`
Ordem interna:
- Mensagem de boas-vindas temática
- Chamar criação de conta
- Chamar escolha de personagem
- Chamar itens iniciais → montar inventário
- Mostrar inventário inicial
- Chamar evento do baú
- Chamar evento do ladrão
- Chamar ficha final


````python
import random 

user_email = str(input("Digite o seu Email: "))
user_password = int(input("Digite a sua Senha Numérica: "))

list_user_input_email = [
    "kaua@gmail.com",
    "vitoriagmail.com",
    "anita@gmailcom",
    "  MARIA@GMAIL.COM  ",
    "lucas@g mail com",
]

list_user_input_email.append(user_email)

list_user_valid_email = []
list_user_invalid_email = []

for email in list_user_input_email:
    valid_email = email.strip().lower().replace(" ", "")
    if "@" not in valid_email or "." not in valid_email or len(valid_email) > 25:
        list_user_invalid_email.append(email)
        is_email_valid = False
    else:
        list_user_valid_email.append(valid_email)
        is_email_valid = True

list_user_input_password = [
    1234,
    5678,
    123,
    123456789
]

list_user_input_password.append(user_password)

list_user_valid_password = []
list_user_invalid_password = []

for password in list_user_input_password:
    if len(str(password)) > 3 and len(str(password)) < 9:
        list_user_valid_password.append(password)
        is_password_valid = True
    else:
        list_user_invalid_password.append(password)
        is_password_valid = False

print(f"-" * 5, "EMAILS E SENHAS CORRETOS", "-" * 5)
for i in list_user_valid_email:
    print(i)
for i in list_user_valid_password:
    print(i)

print(f"-" * 5, "EMAILS E SENHAS INCORRETOS", "-" * 5)
for i in list_user_invalid_email:
    print(i)
for i in list_user_invalid_password:
    print(i)

if is_email_valid == True and is_password_valid == True:
    print("="*30)
    print(f"-" * 5, "INICIANDO O JOGO", "-" * 5,)
    print("="*30)
    
    nomes_proibidos = ["admin", "deus", "mestre"]

while True:
    character_user = input("Digite o nome do seu personagem: ")
    nome = character_user.strip()
    if not nome.replace(" ", "").isalpha():
        print("Nome inválido! Use apenas letras.")
        continue
    if nome.lower() in nomes_proibidos:
        print("Esse nome não é permitido! Escolha outro.")
        continue
    print(f"{character_user} é um nome válido!\n CONTINUE E ESCOLHA SEU TIPO DE PERSONAGEM!")
    break

````
